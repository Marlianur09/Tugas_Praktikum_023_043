```python
import cv2
import numpy as np

training_dir = "/tmp/Skin cancer ISIC The International Skin Imaging Collaboration/Train"
# Gather data train
train_data = []
train_label = []
for r, d, f in os.walk(training_dir):
    for file in f:
        if ".jpg" in file:
            imagePath = os.path.join(r, file)
            image = cv2.imread(imagePath)
            image = cv2.resize(image, (100,100))
            train_data.append(image)
            label = imagePath.split(os.path.sep)[-2]
            train_label.append(label)

train_data = np.array(train_data)
train_label = np.array(train_label)
```

```python
testing_dir = "/tmp/Skin cancer ISIC The International Skin Imaging Collaboration/Test"
# Gather data test
test_data = []
test_label = []
for r, d, f in os.walk(testing_dir):
    for file in f:
        if ".jpg" in file:
            imagePath = os.path.join(r, file)
            image = cv2.imread(imagePath)
            image = cv2.resize(image, (100,100))
            test_data.append(image)
            label = imagePath.split(os.path.sep)[-2]
            test_label.append(label)

test_data = np.array(test_data)
test_label = np.array(test_label)
```

```python
print("Train Data = ", train_data.shape)
print("Train Label = ", train_label.shape)
print("test Data = ", test_data.shape)
print("test Label = ", test_label.shape)
```

```python
print("Data sebelum di-normalisasi ", train_data[0][0][0])

x_train = train_data.astype('float32') / 255.0
x_test = test_data.astype('float32') / 255.0
print("Data setelah di-normalisasi ", x_train[0][0][0])
```

```python
from sklearn.preprocessing import LabelEncoder

print("Label sebelum di-encoder ", train_label[450:480])

lb = LabelEncoder()
y_train = lb.fit_transform(train_label)
y_test = lb.fit_transform(test_label)

print("Label setelah di-encoder ", y_train[450:480])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Conv2D, MaxPool2D, GlobalMaxPooling2D, AveragePooling2D, GlobalAveragePooling2D, Flatten
```

```python
model = Sequential()

model.add(InputLayer(input_shape=[100,100,3]))
model.add(Conv2D(filters=16, kernel_size=3, strides=1, padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=2, padding='same'))
model.add(Conv2D(filters=32, kernel_size=3, strides=1, padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=2, padding='same'))
model.add(GlobalMaxPooling2D(data_format='channels_last'))

model.add(Flatten())
```

```python
# Fully Connected Layer
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
print(model.summary())
```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(lr=0.001), 
              loss='categorical_crossentropy', 
              metrics=['acc'])
```

```python
H = model.fit(x_train, y_train, batch_size=20, epochs=20, validation_data=(x_test, y_test))
```

```python
model2 = Sequential()

model2.add(InputLayer(input_shape=[100,100,3]))
model2.add(Conv2D(filters=16, kernel_size=3, strides=1, padding='same', activation='relu'))
model2.add(AveragePooling2D(pool_size=2, padding='same'))
model2.add(Conv2D(filters=32, kernel_size=3, strides=1, padding='same', activation='relu'))
model2.add(AveragePooling2D(pool_size=2, padding='same'))
model2.add(Conv2D(filters=64, kernel_size=3, strides=1, padding='same', activation='relu'))
model2.add(GlobalAveragePooling2D(data_format='channels_last'))

model2.add(Flatten())
```

```python
# Fully Connected Layer
model2.add(Dense(128, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
```

```python
print(model2.summary())
```

```python
model2.compile(optimizer=Adam(lr=0.001), 
              loss='binary_crossentropy', 
              metrics=['acc'])
```

```python
H2 = model2.fit(x_train, y_train, batch_size=20, epochs=20, validation_data=(x_test, y_test))
```
